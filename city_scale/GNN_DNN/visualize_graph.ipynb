```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
adjacency_matrix_8junc = np.array([[0,1,0,1,0,1,0,0], #1
                                   [1,0,1,1,0,0,0,0], #2
                                   [0,1,0,0,1,0,0,0], #3
                                   [1,1,0,0,1,1,1,0], #4
                                   [0,0,1,1,0,0,0,1], #5
                                   [1,0,0,1,0,0,1,0], #6
                                   [0,0,0,1,0,1,0,1], #7
                                   [0,0,0,0,1,0,1,0]]) #8
```

```python
G = nx.DiGraph(adjacency_matrix_8junc)

labels = [1,2,3,4,5,6,7,8]  # Replace with your actual node labels
for idx, node in enumerate(G.nodes()):
    G.nodes[node]['label'] = labels[idx]
```

```python
nx.draw_networkx(G, labels=nx.get_node_attributes(G, 'label'), with_labels=True)
plt.show()
```
