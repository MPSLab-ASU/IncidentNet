```python
import matplotlib.pyplot as plt

# Segment the data into two parts: outside and inside the range 217000 to 222000
l_outside_range = df_test[df_test['step'] < 216200]
r_outsiide_range = df_test[df_test['step'] > 222000]
inside_range = df_test[(df_test['step'] >= 216200) & (df_test['step'] <= 222000)]

# Plotting the segment outside the range in green
plt.plot(l_outside_range['step'], l_outside_range['rolling_junction_mean_speed_10'], color='green')

# Plotting the segment inside the range in red
plt.plot(inside_range['step'], inside_range['rolling_junction_mean_speed_10'], color='red')
plt.plot(r_outsiide_range['step'], r_outsiide_range['rolling_junction_mean_speed_10'], color='green')

plt.xlabel('Time Step')
plt.ylabel('Mean Junction speed')
plt.title('Junction Speed and Incident Duration')
# Add grid
plt.grid(True)

# Show the plot
plt.show()
```
